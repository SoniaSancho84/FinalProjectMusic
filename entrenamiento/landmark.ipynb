```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os
import numpy as np
from sklearn.svm import SVC
import pickle
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, recall_score, precision_score
```

```python
separator = os.path.sep
path_act = os.path.dirname(os.path.abspath("__file__"))
dir = separator.join(path_act.split(separator)[:-1])
```

```python
def generated_tts():
    df = pd.read_csv(f'{dir}/dataset.csv')
    df.columns = [i for i in range(df.shape[1])]
    df = df.rename(columns={63: 'Output'})
    X = df.iloc[:, :-1]
    Y = df.iloc[:, -1]
    return train_test_split(X, Y, test_size=0.2)
```

```python
def path_act():
    separator = os.path.sep
    path_act = os.path.dirname(os.path.abspath("__file__"))
    dir = separator.join(path_act.split(separator)[:-1])
    return str(dir)
```

```python
X_train, X_test, y_train, y_test = generated_tts()
```

```python
X_train
```

```python
svm = SVC(C=10)
svm.fit(X_train, y_train)
```

```python
y_predict = svm.predict(X_test)
y_predict
```

```python
f1 = f1_score(y_test, y_predict, average='micro')
recall = recall_score(y_test, y_predict, average='micro')
precision = precision_score(y_test, y_predict, average='micro')
f1, recall, precision
```

```python
dir = path_act()
```

```python
with open(f'{dir}/model/modelo.pkl','wb') as fw:
    pickle.dump(svm,fw)
```
